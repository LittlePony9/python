Задание 1
Написать программу «справочник».Создать два списка
целых. Один список хранит идентификационные коды,
второй — телефонные номера. Реализовать меню для
пользователя:
■ Отсортировать по идентификационным кодам;
■ Отсортировать по номерам телефона;
■ Вывести список пользователей с кодами и телефонами;
■ Выход.

```python
def display_users(ids, phones):
    print("\nСписок пользователей:")
    print("{:<15} {:<15}".format("ID-код", "Телефон"))
    print("-" * 30)
    for id_, phone in zip(ids, phones):
        print("{:<15} {:<15}".format(id_, phone))
    print("-" * 30)


def sort_by_ids(ids, phones):
    combined = list(zip(ids, phones))
    combined.sort(key=lambda x: x[0])
    ids[:], phones[:] = zip(*combined)


def sort_by_phones(ids, phones):
    combined = list(zip(ids, phones))
    combined.sort(key=lambda x: x[1])
    ids[:], phones[:] = zip(*combined)


def main():
    ids = [103, 101, 102, 104]
    phones = [789456123, 123456789, 456789123, 987654321]

    while True:
        print("\nМеню:")
        print("1. Отсортировать по идентификационным кодам")
        print("2. Отсортировать по номерам телефона")
        print("3. Вывести список пользователей с кодами и телефонами")
        print("4. Выход")
        choice = input("Выберите действие (1-4): ")

        if choice == "1":
            sort_by_ids(ids, phones)
            print("\nСписок отсортирован по идентификационным кодам.")
        elif choice == "2":
            sort_by_phones(ids, phones)
            print("\nСписок отсортирован по номерам телефона.")
        elif choice == "3":
            display_users(ids, phones)
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный ввод. Попробуйте снова.")

if __name__ == "__main__":
    main()
```

Задание 2
Написать программу «книги». Создать два списка
годы выпуска. Реализовать меню для пользователя:
■ Отсортировать по названию книг;
■ Отсортировать по годам выпуска;
■ Вывести список книг с названиями и годами выпуска;
■ Выход;

```python
def display_books(titles, years):
    print("\nСписок книг:")
    print("{:<30} {:<10}".format("Название книги", "Год выпуска"))
    print("-" * 40)
    for title, year in zip(titles, years):
        print("{:<30} {:<10}".format(title, year))
    print("-" * 40)


def sort_by_titles(titles, years):
    combined = list(zip(titles, years))
    combined.sort(key=lambda x: x[0]) 
    titles[:], years[:] = zip(*combined) 


def sort_by_years(titles, years):
    combined = list(zip(titles, years))
    combined.sort(key=lambda x: x[1])
    titles[:], years[:] = zip(*combined)


def main():
    titles = ["Война и мир", "Преступление и наказание", "Мастер и Маргарита", "1984", "Гарри Поттер"]
    years = [1869, 1866, 1967, 1949, 1997]

    while True:
        print("\nМеню:")
        print("1. Отсортировать по названию книг")
        print("2. Отсортировать по годам выпуска")
        print("3. Вывести список книг с названиями и годами выпуска")
        print("4. Выход")
        choice = input("Выберите действие (1-4): ")

        if choice == "1":
            sort_by_titles(titles, years)
            print("\nСписок отсортирован по названию книг.")
        elif choice == "2":
            sort_by_years(titles, years)
            print("\nСписок отсортирован по годам выпуска.")
        elif choice == "3":
            display_books(titles, years)
        elif choice == "4":
            print("Выход из программы.")
            break
        else:
            print("Некорректный ввод. Попробуйте снова.")


if __name__ == "__main__":
    main()
```
