Задание 1
Показать на экран все простые числа в диапазоне, указанном пользователем. Число называется простым, если оно делится без остатка только на себя и на единицу. Например, три — это простое число, а четыре нет.

```python
start, end = int(input()), int(input())
primes = []

for num in range(start, end + 1):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else: 
            primes.append(num)

print(primes)

```

Задание 2
Показать на экране таблицу умножения для всех чисел
от 1 до 10. Например:
1 * 1 = 1 1 * 2 = 2 ….. 1 * 10 = 10
.........................................................................
10 * 1 = 10 10 * 2 = 20 …. 10 * 10 = 100.

```python
for i in range(1, 11):
    for j in range(1, 11):
        print(f'{i} * {j} = {i * j}')
```

Показать на экран таблицу умножения в диапазоне,
указанном пользователем. Например, если пользователь

3*1 = 3 3*2 = 6 3*3 = 9 ... 3 * 10 = 30
