Задание 1
Создайте программу, хранящую информацию о вели-
ких баскетболистах. Нужно хранить ФИО баскетболиста и
его рост. Требуется реализовать возможность добавления,
удаления, поиска, замены данных. Используйте словарь
для хранения информации.

```python
class BasketballPlayers:
    def __init__(self):
        self.players = {}


    def add_player(self, full_name, height):
        if full_name in self.players:
            print(f"Игрок с именем {full_name} уже существует.")
        else:
            self.players[full_name] = (full_name, height)
            print(f"Игрок {full_name} добавлен.")

    def remove_player(self, full_name):
        if full_name in self.players:
            del self.players[full_name]
            print(f"Игрок {full_name} удален.")
        else:
            print(f"Игрок с именем {full_name} не найден.")

    def search_player(self, full_name):
        if full_name in self.players:
            player = self.players[full_name]
            print(f"Игрок найден: {player[0]}, Рост: {player[1]} см.")
        else:
            print(f"Игрок с именем {full_name} не найден.")

    def update_player(self, full_name, new_full_name, new_height):
        if full_name in self.players:
            self.players[new_full_name] = (new_full_name, new_height)
            if full_name != new_full_name:
                del self.players[full_name]  # Удаляем старое имя
            print(f"Данные игрока {full_name} обновлены.")
        else:
            print(f"Игрок с именем {full_name} не найден.")

    def list_players(self):
        if self.players:
            for player in self.players.values():
                print(f"Имя: {player[0]}, Рост: {player[1]} см.")
        else:
            print("Список игроков пуст.")

basketball_team = BasketballPlayers()

basketball_team.add_player("Michael Jordan", 198)
basketball_team.add_player("LeBron James", 206)

basketball_team.search_player("Michael Jordan")

basketball_team.update_player("Michael Jordan", "Michael Jordan", 199)  # Изменим рост

basketball_team.remove_player("LeBron James")

basketball_team.list_players()
```

Задание 2
Создайте программу «Англо-французский словарь».
Нужно хранить слово на английском языке и его перевод
на французский. Требуется реализовать возможность до-
бавления, удаления, поиска, замены данных. Используйте
словарь для хранения информации.

```python
class EnglishFrenchDictionary:
    def __init__(self):
        self.dictionary = {}

    def add_word(self, english_word, french_translation):
        if english_word in self.dictionary:
            print(f"Слово '{english_word}' уже существует в словаре.")
        else:
            self.dictionary[english_word] = french_translation
            print(f"Слово '{english_word}' с переводом '{french_translation}' добавлено.")

    def remove_word(self, english_word):
        if english_word in self.dictionary:
            del self.dictionary[english_word]
            print(f"Слово '{english_word}' удалено из словаря.")
        else:
            print(f"Слово '{english_word}' не найдено в словаре.")

    def search_word(self, english_word):
        if english_word in self.dictionary:
            print(f"Перевод слова '{english_word}': {self.dictionary[english_word]}")
        else:
            print(f"Слово '{english_word}' не найдено в словаре.")

    def update_word(self, english_word, new_french_translation):
        if english_word in self.dictionary:
            self.dictionary[english_word] = new_french_translation
            print(f"Перевод слова '{english_word}' обновлен на '{new_french_translation}'.")
        else:
            print(f"Слово '{english_word}' не найдено в словаре.")

    def list_words(self):
        if self.dictionary:
            print("Все слова в словаре:")
            for english_word, french_translation in self.dictionary.items():
                print(f"{english_word} -> {french_translation}")
        else:
            print("Словарь пуст.")


dictionary = EnglishFrenchDictionary()


dictionary.add_word("hello", "bonjour")
dictionary.add_word("goodbye", "au revoir")


dictionary.search_word("hello")


dictionary.update_word("hello", "salut")


dictionary.remove_word("goodbye")


dictionary.list_words()
```

Задание 3
Создайте программу «Фирма». Нужно хранить ин-
формацию о человеке: ФИО, телефон, рабочий email,
название должности, номер кабинета, skype. Требуется
реализовать возможность добавления, удаления, поис-
ка, замены данных. Используйте словарь для хранения
информации.

```python
class Company:
    def __init__(self):
        self.employees = {}

    def add_employee(self, full_name, phone, email, position, office_number, skype):
        if full_name in self.employees:
            print(f"Сотрудник с именем {full_name} уже существует.")
        else:
            self.employees[full_name] = {
                'phone': phone,
                'email': email,
                'position': position,
                'office_number': office_number,
                'skype': skype
            }
            print(f"Сотрудник {full_name} добавлен.")

    def remove_employee(self, full_name):
        if full_name in self.employees:
            del self.employees[full_name]
            print(f"Сотрудник {full_name} удален.")
        else:
            print(f"Сотрудник с именем {full_name} не найден.")

    def search_employee(self, full_name):
        if full_name in self.employees:
            employee = self.employees[full_name]
            print(f"Сотрудник найден: {full_name}")
            print(f"Телефон: {employee['phone']}")
            print(f"Email: {employee['email']}")
            print(f"Должность: {employee['position']}")
            print(f"Номер кабинета: {employee['office_number']}")
            print(f"Skype: {employee['skype']}")
        else:
            print(f"Сотрудник с именем {full_name} не найден.")

    def update_employee(self, full_name, phone=None, email=None, position=None, office_number=None, skype=None):
        if full_name in self.employees:
            employee = self.employees[full_name]
            if phone:
                employee['phone'] = phone
            if email:
                employee['email'] = email
            if position:
                employee['position'] = position
            if office_number:
                employee['office_number'] = office_number
            if skype:
                employee['skype'] = skype
            print(f"Данные сотрудника {full_name} обновлены.")
        else:
            print(f"Сотрудник с именем {full_name} не найден.")

    def list_employees(self):
        if self.employees:
            print("Все сотрудники фирмы:")
            for full_name, details in self.employees.items():
                print(f"ФИО: {full_name}")
                print(f"Телефон: {details['phone']}")
                print(f"Email: {details['email']}")
                print(f"Должность: {details['position']}")
                print(f"Номер кабинета: {details['office_number']}")
                print(f"Skype: {details['skype']}")
                print("-" * 20)
        else:
            print("Нет сотрудников в фирме.")

company = Company()

company.add_employee("Иван Иванов", "123-456", "ivanov@company.com", "Менеджер", "101", "ivanov123")
company.add_employee("Мария Петрова", "789-012", "petrova@company.com", "Разработчик", "202", "maria202")

company.search_employee("Иван Иванов")

company.update_employee("Иван Иванов", phone="555-123", email="ivanov_new@company.com")

company.remove_employee("Мария Петрова")

company.list_employees()
```

Задание 4
Создайте программу «Книжная коллекция». Нужно
хранить информацию о книгах: автор, название книги,
жанр, год выпуска, количество страниц, издательство.
Требуется реализовать возможность добавления, удале-
ния, поиска, замены данных. Используйте словарь для
хранения информации.

```python
class BookCollection:
    def __init__(self):
        self.books = {}

    def add_book(self, title, author, genre, year, pages, publisher):
        if title in self.books:
            print(f"Книга с названием {title} уже существует в коллекции.")
        else:
            self.books[title] = {
                'author': author,
                'genre': genre,
                'year': year,
                'pages': pages,
                'publisher': publisher
            }
            print(f"Книга '{title}' добавлена в коллекцию.")

    def remove_book(self, title):
        if title in self.books:
            del self.books[title]
            print(f"Книга '{title}' удалена из коллекции.")
        else:
            print(f"Книга с названием {title} не найдена в коллекции.")

    def search_book(self, title):
        if title in self.books:
            book = self.books[title]
            print(f"Информация о книге '{title}':")
            print(f"Автор: {book['author']}")
            print(f"Жанр: {book['genre']}")
            print(f"Год выпуска: {book['year']}")
            print(f"Количество страниц: {book['pages']}")
            print(f"Издательство: {book['publisher']}")
        else:
            print(f"Книга с названием {title} не найдена в коллекции.")

    def update_book(self, title, author=None, genre=None, year=None, pages=None, publisher=None):
        if title in self.books:
            book = self.books[title]
            if author:
                book['author'] = author
            if genre:
                book['genre'] = genre
            if year:
                book['year'] = year
            if pages:
                book['pages'] = pages
            if publisher:
                book['publisher'] = publisher
            print(f"Данные книги '{title}' обновлены.")
        else:
            print(f"Книга с названием {title} не найдена в коллекции.")

    def list_books(self):
        if self.books:
            print("Все книги в коллекции:")
            for title, details in self.books.items():
                print(f"Название: {title}")
                print(f"Автор: {details['author']}")
                print(f"Жанр: {details['genre']}")
                print(f"Год выпуска: {details['year']}")
                print(f"Количество страниц: {details['pages']}")
                print(f"Издательство: {details['publisher']}")
                print("-" * 20)
        else:
            print("Коллекция книг пуста.")

collection = BookCollection()

collection.add_book("Война и мир", "Лев Толстой", "Роман", 1869, 1200, "Эксмо")
collection.add_book("1984", "Джордж Оруэлл", "Дистопия", 1949, 328, "Росмэн")

collection.search_book("Война и мир")

collection.update_book("1984", pages=350, publisher="АСТ")

collection.remove_book("Война и мир")

collection.list_books()
```
